## グラフの描画

```python
# 可視化するデータの準備
X = np.linspace(0, 4*np.pi, 1000)
Y = np.sin(X)
```

```python
# 中身の一部はこんな感じ
print(X[:5])
print(Y[:5])
```

```python
# 描画
fig, ax = plt.subplots()
ax.plot(X, Y)
fig.show()
```

```python
# 描画用の関数作成（2変数）
def plot(X, Y, mode="plot"):
    fig, ax = plt.subplots()
    if mode == "plot":
        ax.plot(X, Y)
    if mode == "scatter":
        ax.scatter(X, Y)
    if mode == "bar":
        ax.bar(X, Y)
    if mode == "errorbar":
        ax.errorbar(X, Y, Y/4)
    fig.show()
```

```python
# 描画用の関数作成（1変数）
def plot2(Z,  mode="imshow"):
    fig, ax = plt.subplots()
    if mode == "imshow":
        ax.imshow(Z)
    if mode == "contourf":
        ax.contourf(Z)
    if mode == "pie":
        ax.pie(Z)
    if mode == "hist":
        ax.hist(Z)
    if mode =="boxplot":
        ax.boxplot(Z)
    fig.show()
```

```python
# 散布図
X = np.random.uniform(0, 1, 100)
Y = np.random.uniform(0, 1, 100)
plot(X, Y, mode="scatter")
```

```python
# 棒グラフ
X = np.arange(10)
Y = np.random.uniform(1, 10, 10)
plot(X, Y, "bar")
```

## グラフの調整

```python
# sin waveを弄ってみる
def sinw(color="", linestyle="", linewidth="", marker=""):
    X = np.linspace(0, 10, 100)
    Y = np.sin(X)
    fig, ax = plt.subplots()
    ax.plot(X, Y)
    if color != "":
        fig.gca().get_lines()[0].set_color(color)
    if linestyle != "":
        fig.gca().get_lines()[0].set_linestyle(linestyle)
    if linewidth != "":
        fig.gca().get_lines()[0].set_linewidth(linewidth)
    if marker != "":
        fig.gca().get_lines()[0].set_marker(marker)
    fig.show()
```

```python
sinw()
```

```python
sinw(color="black")
```

```python
sinw(linestyle="--")
```

```python
sinw(linewidth=5)
```

```python
sinw(marker="o")
```

## グラフの整理

```python
# 重ねて描画
X = np.linspace(0, 10, 100)
Y1, Y2 = np.sin(X), np.cos(X)
fig, ax = plt.subplots()
ax.plot(X, Y1)
ax.plot(X, Y2)
fig.show()
```

```python
# 上下に分割
X = np.linspace(0, 10, 100)
Y1, Y2 = np.sin(X), np.cos(X)
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)
ax1.plot(X, Y1, color="blue")
ax2.plot(X, Y2, color="orange")
fig.show()
```

```python
# 左右に分割
X = np.linspace(0, 10, 100)
Y1, Y2 = np.sin(X), np.cos(X)
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
ax1.plot(Y1, X, color="blue")
ax2.plot(Y2, X, color="orange")
fig.show()
```

## ラベル

```python
X = np.linspace(0, 10, 100)
Y = np.sin(X)
fig, ax = plt.subplots()
ax.plot(X, Y)
fig.suptitle(None)
ax.set_title("A Sine Wave")
fig.show()
```
